```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = [1,2] #weights 
b = 3
```

```python
x = np.linspace(-5,5,100)
y = w[0]*x + w[1]*x + b
```

```python
plt.plot(x,y, label='Linear Combination')
plt.legend()
plt.show()
```

```python
sigmoid = lambda x: 1/(1+np.exp(-x))
y_sigmoid = sigmoid(y)
plt.plot(x,y_sigmoid,label='sigmoid function')
plt.legend()
plt.show()
```

```python
relu = lambda x:np.maximum(0,x)
y_relu = relu(y)
plt.plot(x,y_relu,label='RELU ACTIVATION')
plt.legend()
plt.show()
```
