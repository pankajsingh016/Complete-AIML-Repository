```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
plt.style.use('fivethirtyeight')
```

```python
x = np.linspace(-10,10,100)

def plot_graph(x,y,ALPHA=0.6, label_x=r"$x \rightarrow$",label_y=r"$act(x) \rightarrow$",title=None,LABEL=None):
    plt.figure(figsize=(7,5))
    plt.axhline(y=0, color="black", linestyle="--",lw=2)
    plt.axvline(x=0, color="black", linestyle="--",lw=2)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.title(title)

    if LABEL != None:
        plt.plot(x,y,alpha=ALPHA,label=LABEL)
        plt.legend(fontsize=14)
    
    else:
        plt.plot(x,y,alpha=ALPHA)

def derivative(f,z,delta_z = 1e-6):
    return (f(z+delta_z) - f(z))/(delta_z)
```

```python
def sigmoid(x):
    return tf.keras.activations.sigmoid(x)
```

```python
plot_graph(x,sigmoid(x),title="Sigmoid function",LABEL=r"$\sigma(x)= \frac{1}{1+e^{-x}}$")
```

```python
plot_graph(x,derivative(sigmoid,x),title="Derivative of sigmoid functions", LABEL=r"$\sigma^\prime(x)=\sigma(x)(1-\sigma(x))$")
```

```python
def tanh(x):
    return tf.keras.activations.tanh(x)
```

```python
plot_graph(x,tanh(x),title="Hyperbolic Tangent",LABEL=r"$tanh(x) = \frac{(e^{x} - e^{-x})}{(e^{x} + e^{-x})}$")
```

```python
plot_graph(x,derivative(tanh,x),title="Derivative of tanh function",LABEL=r"$tanh^\prime(x) = 1-tanh^2(x)$")
```

```python
# Leaky RELU

def leaky_relu(x,alpha=0.3):
    return np.where(x >=0, x, alpha*x)
```

```python
plot_graph(x,leaky_relu(x),title="Leaky Relu")
```

```python
plot_graph(x,derivative(leaky_relu,x),title="Derivative of LeakyRelu function",LABEL=r"$Leaky\_ReLU^\prime(x)$")
```

```python
#softmax

tf.keras.activations.softmax(tf.constant([[0.3,0.3,0.4,0.5,0.7]]),axis=1)
```

```python
# Hard Sigmoid
# Hard coded-values for different value

def hard_sigmoid(z):
    z = tf.constant(z)
    return tf.keras.activations.hard_sigmoid(z)
```
