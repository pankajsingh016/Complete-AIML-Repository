## Standardization

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill = df['total_bill']
```

```python
import numpy as np
mean = np.mean(total_bill)
```

```python
std = np.std(total_bill)
```

```python
norm_data = []
for i in total_bill:
    z_score = (i-mean)/std
    norm_data.append(z_score)
```

```python
sns.histplot(total_bill)
```

```python
sns.histplot(norm_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
import pandas as pd
scaled_df_col = pd.DataFrame(scaler.transform(df[['total_bill','tip']]),columns=['total_bill','tip'])
```

```python
scaled_df_col.head()
```

```python
sns.histplot(scaled_df_col['total_bill'])
```

```python
sns.histplot(scaled_df_col['tip'])
```
