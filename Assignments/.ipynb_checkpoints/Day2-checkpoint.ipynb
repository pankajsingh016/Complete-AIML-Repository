```python
import numpy as np
import pandas as pd 
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

from numpy.random import randn,randint,uniform,sample
```

```python
df = pd.DataFrame(randn(10,4),columns=['a','b','c','d'])
```

```python
df
```

```python
df.plot(kind='bar',stacked=False,figsize=(10,10))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar chart (unstacked)')
```

```python
df.plot(kind='bar',stacked=True,figsize=(10,10))
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar chart (stacked)')
```
