#### **What is Plotly?**
- Plotly is a Python graphing Library. It Provides an interactive way to display the data using publication-quality graphs.

- It is open source, provides wide variety of graph like scatter,line plots, area,pie,bar charts, error bars, box plots, histogram, heatmaps, subplots,multiple-axes, polar charts, and bubble charts.

### **Where it is Used?**
Ploty can be used in plotting various graphs like
1. Basic charts (Schools)
2. Statistical Charts (Company or research)
3. Scientific Charts (Research and Experiments)
4. Financial Charts (Banks or economical survey)
5. Maps (Bubble maps for any region)
6. 3D charts (To visualize 3 variables)
7. Sub-Plots
8. Animated charts



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

### 1.India's Best Selling Car Market Share % values:



```python
# Best Selling Cars and something about them
car = pd.read_csv("/content/Car Data.csv")
car = car.loc[0:9]
car.head()
```

```python
#Normal Pie Chart
import plotly.express as px
fig = px.pie(car,values='Sales Per Unit',names='CARS')
fig.show()
```

```python
# Pie Chart with hole in it
import plotly.graph_objects as go
labels = list(car['CARS'])
values = list(car['Sales Per Unit'])
fig2 = go.Figure(data=[go.Pie(labels=labels,values=values,hole=.4)])
fig2.show()
```

### 2. Bar Charts with India Covid 19 Active Cases

```python
# Plot a Bar chart with India covid 19 Active cases
covid = pd.read_excel(r"/content/Covid Data.xlsx")
covid = covid.loc[0:10]
covid.head()
```

```python
fig3 = px.bar(covid,x='Location',y='Cases',title="India's States and Covid-19 Active Cases",labels={'Cases':'No. of Cases','Location':'States'})
fig3.show()
```

```python
fig4 = px.bar(covid,x='Location',y='Cases',hover_data=['Cases','Recovered'],color='Cases',labels={'Cases':'No. of Cases','Location':'States'},title="India's States and Covid-19 Active Cases")
fig4.show()
```

### 3.Line Chart of Covid-19 Recovered Cases

```python
# Plot a Line chart with India covid-19 recovery cases
Recovery = pd.read_excel(r"/content/Monthly Data.xlsx")
Recovery.head()
```

```python
fig5 = px.line(Recovery,x='Month ',y='Recovered',title="India's Covid-19 Recovered Cases 2020")
fig5.show()
```
