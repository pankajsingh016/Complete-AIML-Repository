## Color Thresholding

Color Thresholding is a technique to filter out range of colors in an image by defining upper and lower bounds in a color  space, We use it to segment, replace or isolate specific colors in an image.

- It require a color space
  
**RGB:** Affects all the channel 
**HSV:** It is better for color targeting as color is differnt from brightness
**H** [0, 360] degree
**S** [0, 100] percentage
**V** [0, 100] percentage

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('img_1.jpg')
```

```python
plt.imshow(image)
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img_rgb)
```

```python
plt.imshow(img_hsv)
```

```python
lower_white = np.array([0,0,190])
upper_white = np.array([30,80,255])
```

```python
mask = cv2.inRange(img_hsv, lower_white, upper_white)
```

```python
plt.imshow(mask)
```

```python
blue_hsv = np.uint8([[[120,75, 100]]])
blue_bgr = cv2.cvtColor(blue_hsv, cv2.COLOR_HSV2BGR)[0][0]
```

```python
image_result = image.copy()
image_result[mask > 0 ] = blue_bgr
```

```python
image_result_rgb = cv2.cvtColor(image_result, cv2.COLOR_BGR2RGB)
```

```python
# PLOTS
fig, axs = plt.subplots(1,3,figsize=(12,6))
axs[0].imshow(img_rgb)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(mask)
axs[1].set_title('Binary Mask')
axs[1].axis('off')

axs[2].imshow(image_result_rgb)
axs[2].set_title('Output Image')
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
