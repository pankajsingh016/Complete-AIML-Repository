```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Creating the canvas
height, width = 500,700
blue = (255, 126,0)
canvas = np.full((height,width,3),blue, dtype=np.uint8)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

### Draw a Line

```python
# (canvas, (starting coordinates),(ending coordinates),(color coordinates), thicknesss)
cv2.line(canvas,(50,50),(650,450),(255,255,255),2)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw a Rectangle

rectangle(canvas (starting coordinates),(ending coordinates),(color),fill type)

fill-type > 0 (thickness on border)
fill-type < 0 (compelete fill the box)

```python
cv2.rectangle(canvas,(400,400),(300,300),(0,255,255),-1)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw a circle

Circle(canvas, (starting point coordinates),radious,(coor),fill-type)

```python
cv2.circle(canvas,(500,150),100,(0,0,255),5)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw an Ellipse
Ellipse(canvas, (starting coordinates),(major axis, minor axis),elliplse angle,start angle, end angle, (color), fill-type)

```python
cv2.ellipse(canvas,(100,300),(100,140),45,0,360,(0,255,0),3)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

### Draw a polygon
Xn, Yn, where n > 2 or n>3 or n>10 even

polyline(canvas , [your connecting points],True/False(closed or not),color, thickness)

```python
points = np.array([[150,400],[300,400],[225,300]],np.int32)
new_reshape_points = points.reshape((-1,1,2))
# new_reshape_points

cv2.polylines(canvas , [new_reshape_points], True, (255,0,255),3)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```
