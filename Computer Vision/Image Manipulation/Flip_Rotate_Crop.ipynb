```python
import cv2 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./images/img_3.jpg")
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

### FLIP IMAGE
- 1 ---> Horizontal Flip
- 0 ---> Vertical Flip
- -ve 1 ---> Both Axes

```python
flip_horizontal = cv2.flip(img_rgb,1)
flip_vertical = cv2.flip(img_rgb, 0)
flip_both = cv2.flip(img_rgb,-1)

titles = ['Original image', 'Horizontal Flip','Vertical Flip', 'Both Axes']
images = [img_rgb, flip_horizontal, flip_vertical, flip_both]
plt.figure(figsize=(8,8))
for i in range(len(images)):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

### Rotate


```python
rotate_90_clockwise = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterClock = cv2.rotate(img_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(img_rgb, cv2.ROTATE_180)
rotate_270 = cv2.rotate(rotate_180, cv2.ROTATE_90_CLOCKWISE) ## Roate the image to 270 degrees

title = ['original image','90 degree rotated', '180 degree rotated', '270 degree rotated','90 degree counter clock']
images = [
    img_rgb,
    rotate_90_clockwise,
    rotate_180,
    rotate_270,
    rotate_90_counterClock
]

plt.figure(figsize=(8,8))

for i in range(len(images)):
    plt.subplot(3,3,i+1)
    plt.imshow(images[i])
    plt.title(title[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
