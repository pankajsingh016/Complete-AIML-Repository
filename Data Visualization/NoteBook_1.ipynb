## What Chart to use when?

### Line chart
- Uses points to display the data.
- when? you have to disply the data over a period of time.
 
### Bar chart
- Comparing different variables
- Use to show the changes
- Men vs Women
- pattern studies behavioural studies.

### Scatter plot
- Relationship between any two variables.
- population rate vs child motality rate.

### Pie chart
- Shows percentage (each and every slice)
- Employement percentage in india.
- Player performance

#### Line Plot

```python
## Simple Example
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.plot(x,y,'go--', linewidth=2, markersize=12)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
x = np.arange(1,11)
y = 3*x + 5
plt.title("Line plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,'go--') # find out more colors in plotting
```

```python
# Compute the x and y corrdinates for points on a sine curve
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("Sine Waveform")

# Plot the points using matplotblib
plt.plot(x,y,color='g')
plt.show()
```

```python
x = np.arange(0,10)
y = x*x
plt.title("Parabola")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y,color='b')
```

```python
## Assignment 1
x = np.arange(1,10)
y = 3*x + 10
## plt plot
plt.plot(x,y, 'ro--')
#pt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize(12))
plt.xlabel('x-axis',color='blue')
plt.ylabel('y-axis',color= 'green')
```

```python
## creating subplots

plt.subplot(2,2,1)  # (row=2,column=2,segment=1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)  #(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)  #(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)  #(2,2,4)
plt.plot(x,y,'go')
```
