```python
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
# total_bill and tip is a numerical dataset
# sex is a categorical variable 
# smoker,day,time,size -> categorical variable

tips['size']
```

```python
# relational plot
sns.relplot(x='total_bill',y='tip',data=tips)
plt.grid()
```

```python
from numpy.random import randn
```
