```python
import pandas as pd
```

```python
df = pd.read_csv("..\\data\\spam.csv",encoding='latin-1')
df = df.iloc[:,:2]
df.columns = ['label','message']
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

corpus_porter_stemmer = []
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ',df['message'][i])
    review = review.lower()
    review = review.split(' ')
    review = [ ps.stem(word) for word in review if not word in  stopwords.words('english')]
    review = ' '.join(review)
    corpus_porter_stemmer.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True)
# max_feature is the size of the vocublary
# for binary bag of word enable binary = True
X = cv.fit_transform(corpus_porter_stemmer).toarray()
```

```python
#Ngram range
cv = CountVectorizer(max_features=500,binary=True,ngram_range=(2,3))
# max_feature is the size of the vocublary
# for binary bag of word enable binary = True
X = cv.fit_transform(corpus_porter_stemmer).toarray()
```
