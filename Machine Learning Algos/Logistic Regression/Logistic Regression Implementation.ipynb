```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset.keys()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df.head()
```

## Appending target to the dataset

```python
df['target'] = dataset.target
```

```python
df.head()
```

## Binary Classification

```python
df_bin_cl = df[df['target']!=2]
```

```python
df_bin_cl.head()
```

## Independent and dependent feature

```python
X = df_bin_cl.iloc[:,:-1]
y = df_bin_cl.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
classifier.fit(X_train,y_train)
```

```python
## Prediction
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
##Confusion Matrix, Accuracy Score, Classification report
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_pred,y_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```

### HyperParameter Tuning

##### Grid Search Cross Validation

```python
##Gridsearch CV
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
parameters = {'penalty':('l1','l2','elasticnet',None),'C':[1,10]}
```

```python
clf = GridSearchCV(classifier,param_grid=parameters,cv=5)
```

```python
## Splitting of train data to validation data
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
classifier = LogisticRegression(C=1,penalty='l2')
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
```

## Multi-Class Classification
- One versue Rest
- Multinomial

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

#define dataset
X,y = make_classification(n_samples = 1000, n_features=10, n_informative=5, n_redundant=5, n_classes=3,random_state=42) 
```

```python
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_prob_pred = model.predict_proba(X_test)
```

```python
y_prob_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score, classification_report
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
