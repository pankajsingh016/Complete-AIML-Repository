```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_foreset_fires_clean_data.csv')
```

```python
df.head()
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.columns
```

```python
X = df.drop(['FWI'],axis=1)
y = df['FWI']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.85,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
def correlation(data,threshold):
    col_cor = set()
    correlation_matrix = data.corr()
    for i in range(len(correlation_matrix)):
        for j in range(i):
            if abs(correlation_matrix.iloc[i,j]) > threshold:
                col_cor.add(correlation_matrix.columns[i])

    return col_cor
```

```python
corr_features = correlation(X_train,0.85)
```

```python
corr_features
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

### Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

### Box Plots To Understand Effect of Standard Scaler

```python
plt.subplots(figsize=(15,5))

plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('Before Scaling X_train')

plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('After Scaling X_train')
```

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

LR = LinearRegression()
LR.fit(X_train_scaled,y_train)
y_pred = LR.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)

print("Mean Absolute Error {}".format(mae))
print("R2 Score {}".format(score))
```

### Lasso Regression Model

```python
from sklearn.linear_model import Lasso

L1_Reg = Lasso()
L1_Reg.fit(X_train_scaled,y_train)
y_pred = L1_Reg.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)

print("Mean Absolute Error {}".format(mae))
print("R2 Score {}".format(score))
```

### Ridge Regression Model

```python
from sklearn.linear_model import Ridge

L2_Reg = Ridge()
L2_Reg.fit(X_train_scaled,y_train)
y_pred = L2_Reg.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)

print("Mean Absolute Error {}".format(mae))
print("R2 Score {}".format(score))
```
