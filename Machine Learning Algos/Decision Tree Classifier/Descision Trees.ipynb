```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset.keys()
```

```python
dataset.feature_names
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
df.head()
```

```python
dataset.target
```

```python
X = df.iloc[:,:-1]
y = dataset.target
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## Post Pruning
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train, y_train)
```

```python
def make_decision_tree(clf,x,y):
    from sklearn import tree
    plt.figure(figsize=(x,y))
    tree.plot_tree(clf,filled=True)
    plt.show()
```

```python
make_decision_tree(treeclassifier,8,12)
```

### Post Pruning with Max Depth=2

```python
treeclassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier.fit(X_train,y_train)
```

```python
make_decision_tree(treeclassifier,5,5)
```

### Prediction

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```

## Breast Cancer Prediction Using Decision Trees

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset.keys()
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
df.head().T
```

```python
df.shape
```

```python
dataset.target
```

```python
X,y = dataset.data, dataset.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train,y_train)
```

```python
make_decision_tree(classifier,14,15)
```

### Post Pruning Max_depth = 4 

```python
classifier = DecisionTreeClassifier(max_depth=3)
```

```python
classifier.fit(X_train,y_train)
```

```python
make_decision_tree(classifier,10,10)
```

```python
y_pred = classifier.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```
