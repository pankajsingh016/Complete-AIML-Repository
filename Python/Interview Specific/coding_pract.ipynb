```python
## Fibonacci series printing

def fibo(num):
    a = 0
    b = 1
    l = []
    # 0 1 1 2 3 5 8 .......
    while(num > 0):
        print(a,end=" ")
        l.append(a)
        t = a 
        a = b 
        b = a+t
        num -= 1
    # 0 1 1 
    return l

L = fibo(100)
```

```python
import matplotlib.pyplot as plt
plt.plot(L)
```
