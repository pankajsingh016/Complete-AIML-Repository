```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'].str.split(" ").str[0]
```

```python
df['Arrival_hours'] = df['Arrival_Time'].str.split(" ").str[0].str.split(":").str[0].astype(int)
df['Arrival_minutes'] = df['Arrival_Time'].str.split(" ").str[0].str.split(":").str[1].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_hours'] = df['Dep_Time'].str.split(":").str[0].astype(int)
df['Dep_minute'] = df['Dep_Time'].str.split(":").str[1].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df['Duration'][1].split(" ")
```

```python
for i in range(len(df['Duration'])):
    d = df['Duration'][i]
    el = d.split(" ")
    if len(el) == 1:
        if 'h' in el[0]:
            df['Duration'][i] = df['Duration'][i] + ' 0m'
        if 'm' in el[0]:
            df['Duration'][i] = '0h ' + df['Duration'][i]
```

```python
df['Duration'].str.split(" ")
```

```python
df['Duration_hours'] = df['Duration'].str.split(" ").str[0].str.strip("h").astype(int)
```

```python
df['Duration_minutes'] = df['Duration'].str.split(" ").str[1].str.strip("m").astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```
