# EDA Forest Fire Dataset
We noticed that the data contains two region dataset. So at first we loaded the data in excel and then divided the data into 2 different sheet according to the region as given.

After that we will load the dataset below and concatinate both the dataset index wise.

### Data Loading and Cleaning

```python
import pandas as pd
```

```python
df_Bejaia = pd.read_excel('Algerian_forest_fires_dataset_UPDATE.xlsx',sheet_name='Bejaia')
```

```python
df_Bejaia.head()
```

```python
df_Bejaia.shape
```

```python
df_SidiBelAbbes = pd.read_excel('Algerian_forest_fires_dataset_UPDATE.xlsx',sheet_name='SidiBelAbbes')
```

```python
df_SidiBelAbbes.head()
```

```python
df_SidiBelAbbes.shape
```

```python
df_Bejaia['Region'] = 1
df_SidiBelAbbes['Region'] = 2
```

```python
df_Bejaia.head()
```

```python
df = pd.concat([df_Bejaia,df_SidiBelAbbes],axis=0)
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

### Data Exploration

```python
df.describe()
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.Classes.value_counts()
```

```python
df.Classes = df.Classes.str.strip()
```

```python
df.Classes.value_counts()
```

### Applying One-Hot-Encoding to see the relationship of Classes feature with other features

```python
encoding = {'fire':1,'not fire':0}
df['Classes'] = df['Classes'].replace(encoding)
```

```python
df.Classes.unique()
```

```python
df.Classes.value_counts()
```

```python
df.head()
```

```python
df.drop(df[['day','year']],axis=1,inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.heatmap(df.corr(),annot=True)
plt.title("Correlation between different variables")
plt.show()
```

```python
df.head()
```

```python
fig,axis = plt.subplots(4,3,figsize=(20,18))
# Plotting in each subplot
sns.histplot(data=df, x='Temperature', hue='Classes', ax=axis[0, 0])
sns.histplot(data=df, x='RH', hue='Classes', ax=axis[0, 1])
sns.histplot(data=df, x='Ws', hue='Classes', ax=axis[0, 2])
sns.histplot(data=df, x='Rain', hue='Classes', ax=axis[1, 0])
sns.histplot(data=df, x='FFMC', hue='Classes', ax=axis[1, 1])
sns.histplot(data=df, x='DMC', hue='Classes', ax=axis[1, 2])
sns.histplot(data=df, x='DC', hue='Classes', ax=axis[2, 0])
sns.histplot(data=df, x='ISI', hue='Classes', ax=axis[2, 1])
sns.histplot(data=df, x='BUI', hue='Classes', ax=axis[2, 2])
sns.histplot(data=df, x='FWI', hue='Classes', ax=axis[3, 0])
sns.histplot(data=df, x='Region', hue='Classes',ax=axis[3,1])
sns.histplot(data=df, ax=axis[3,2])

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
df.groupby(['Classes'])['Region'].value_counts()
```

```python
distribution = df.Classes.value_counts(normalize=True)*100
```

```python
plt.title("Fire vs Non Fire Areas")
labels = ['Not Fire', 'Fire']
plt.pie(distribution,labels=labels,autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
df.loc[df['Region']==1]
```

```python
plt.subplots(figsize=(10,4))
df_Bejaia = df.loc[df['Region']==1]
sns.countplot(x="month",hue="Classes",data=df_Bejaia)
plt.legend(labels=['Not fire','Fire'])
plt.xticks(np.arange(4),['June','July','August','September'])
plt.show()
```

```python
plt.subplots(figsize=(10,4))
SidiBelAbbes = df.loc[df['Region']==2]
sns.countplot(x="month",hue="Classes",data=SidiBelAbbes)
plt.legend(labels=['Not fire','Fire'])
plt.xticks(np.arange(4),['June','July','August','September'])
plt.show()
```
